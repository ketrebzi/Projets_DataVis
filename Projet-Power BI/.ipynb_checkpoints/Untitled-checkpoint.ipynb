```python
import pandas as pd

# Charger la feuille 2018 du classeur impots

feuille = '2018'
df = pd.read_excel('C:/Users/ketre/Desktop/Projet-Power BI/impots.xlsx', sheet_name=feuille, header=7)

# Afficher les premières lignes du DataFrame pour vérifier le chargement
print(df.head())
```

```python
import pandas as pd

# Charger la feuille 1984_1999 du classeur impots

feuille = '1984_1999'
df = pd.read_excel('C:/Users/ketre/Desktop/Projet-Power BI/impots.xlsx', sheet_name=feuille, header=7)

# Afficher les premières lignes du DataFrame pour vérifier le chargement
print(df.head())
```

```python
# Supprimer les colonnes où toutes les valeurs sont NaN
df_sans_nans_all = df.dropna(axis=1, how='all')

# Afficher le DataFrame après suppression des colonnes où toutes les valeurs sont NaN
print(df_sans_nans_all.head())
```

```python
# Liste des noms de villes à conserver
villes_a_garder = ['Bouches-du-Rhône', 'Charente', 'Manche', 'Pas-de-Calais', 'Pyrénées-Atlantiques', 'Haute-Savoie', 'Vienne', "Val-D'OIS"]

# Filtrer le DataFrame pour ne conserver que les lignes où la colonne "Nom" contient les villes spécifiées
df_filtre = df[df['Nom'].isin(villes_a_garder)]

# Afficher le DataFrame filtré
print(df_filtre)
```

```python
# Enregistrer le DataFrame filtré dans un fichier Excel
df_filtre.to_excel('C:/Users/ketre/Desktop/Projet-Power BI/classeur_filtre.xlsx', index=False)
```
